# 2.3

```python
π = np.pi

K = 10_000

x = np.random.rand(K)
y = np.random.rand(K)
z = np.sqrt(1.0 - x**2-y**2)

in_circle = (x**2 + y**2) < 1.0
n_in = in_circle.sum()

z[in_circle].sum()/n_in*2*π
```

```python
plt.scatter(x[in_circle][:1000], y[in_circle][:1000])
```
