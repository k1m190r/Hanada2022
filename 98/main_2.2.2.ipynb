# 2.2.2

$$
\lim_{K→∞} 1/K ∑_{k=1}^K f(x^{(k)}) = \frac{1}{b-a} \int_a^b{f(x)dx} \tag{2.3}
$$

$ f(x)=\sqrt{1-x^2} $ and $a=0,b=1$ so $(2.3)=π/4$

```python
np.pi/4
```

```python
K = 1_000
x = np.random.rand(K)
f = np.sqrt(1-x**2)

f.sum()/K
```
