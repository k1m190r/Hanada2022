# Box-Muller Method

$ p,q ‚àà ùí∞(0,1)$

$$
x = \sqrt{-2 log(p)} cos(2œÄq)\\
y = \sqrt{-2 log(p)} sin(2œÄq) 
\tag{2.10}
$$

```python
N = 10_000
p = np.random.rand(N)
q = np.random.rand(N)
```

```python
nx = np.sqrt(-2*np.log(p)) * np.cos(2*np.pi*q)
ny = np.sqrt(-2*np.log(p)) * np.sin(2*np.pi*q)
```

```python
fig, ax = plt.subplots(1,1)
ax.scatter(nx, ny, s=1, alpha=0.1)
```

```python
fig, ax = plt.subplots(1,1)
ax.hist(nx, 100);
```

```python
nx.mean(), nx.std()
```
